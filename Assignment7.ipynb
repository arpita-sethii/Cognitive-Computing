```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
sales_data = np.random.randint(1000, 5001, size=(12, 4))
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)
```

```python
print(sales_df.head())
print(sales_df.describe())
```

```python
sales_df['Total Sales'] = sales_df.sum(axis=1)
total_sales_category = sales_df.sum()

total_sales_month = sales_df['Total Sales']

print(total_sales_category)

print(total_sales_month)
```

```python
sales_growth = sales_df[categories].pct_change().mean()
print(sales_growth)
```

```python
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100
print(sales_df)
```

```python
sales_df['Electronics'] = sales_df['Electronics'] * 0.9
print(sales_df)
```

```python
sales_df[categories].plot(marker='o', linestyle='-')
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution Per Category")
plt.show()
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_values = np.abs(array)
print(abs_values)
percentiles = np.percentile(array, [25, 50, 75], axis=None)
print(percentiles)
mean_val = np.mean(array)
median_val = np.median(array)
std_dev = np.std(array)
print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_dev)
```

```python
b = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_values = np.floor(b)
ceiling_values = np.ceil(b)
truncated_values = np.trunc(b)
rounded_values = np.round(b)
```

```python
print(floor_values)
print(ceiling_values)
print(truncated_values)
print(rounded_values)
```
