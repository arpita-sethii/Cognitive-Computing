```python
import numpy as np
import matplotlib.pyplot as plt

#1
gfg= np.matrix('[4,1,9;12,3,1;4,5,6]')
print(np.sum(gfg))
print("Sum column-wise:",np.sum(gfg,axis=0))
print("Sum row-wise",np.sum(gfg,axis=1))

#2
```

```python
array=np.array([10,52,62,16,16,54,453])
sorted_array=np.sort(array)
sorted_indices=np.argsort(array)
print(sorted_array)
print(sorted_indices)
print(sorted_array[:4])
print(sorted_array[-5:])
```

```python
array_b= np.array([1.0,1.2,2.2,2.0,3.0,2.0])
integer_el=array_b[array_b==array_b.astype(int)]
float_el=array_b[array_b!=array_b.astype(int)]
print(integer_el)
print(float_el)
```

```python
X=ord('A')+ord('I')
sales = np.array([X, X+50, X+100, X+150, X+200])
tax_rate=((X%5)+5)/100
taxed_sales=sales*(1+tax_rate)
d_sales=np.where(sales<X+100, sales*0.95,sales*0.9)
print(taxed_sales)
print(d_sales)
sales_matrix=np.vstack([sales,sales*1.02,sales*1.04])
print(sales_matrix)
```

```python
x=np.linspace(-10,10,100)
y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)

plt.figure(figsize=(10,6))
plt.plot(x,y1,label='y=x^2')
plt.plot(x, y2, label='y = sin(x)')
plt.plot(x, y3, label='y = exp(x)')
plt.plot(x, y4, label='y = log(|x| + 1)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Various Functions using NumPy')
plt.legend()
plt.grid()
plt.show()
```
